```python
# creating database containing non-number values
import pandas as pd
import numpy as np
df = pd.DataFrame({'A':[11,21, np.nan], 'B':[15, np.nan, np.nan], 'C':[10,20,30]})
df
```

```python
#Boolean function
df.isna ()   ##'isna' command's output is true when given number is NaN and false when it IS a number
```

```python
# Inverse of 'isna' command
df.notna ()
```

```python
#columnwise na values
df.isna().sum()
```

```python
#NaN Rows are dropped/deleted--->complete row is removed if it contains even one NaN value
df.dropna ()
```

```python
#To replace NaN values by given value
df.fillna(value=7.5)
```

```python
#Mean values of columns are calculated and then replaced with all NaN values
df.fillna(df.mean())
```

```python
#Step 1--->New dataset
import pandas as pd

data = {
    "Roll No": [101, 102, 103, 104, 105],
    "Name": ["Asha", "Rohan", "Meera", "Karan", "Anil"],
    "Age": [20, "-", 21, 20, "22"],
    "Marks": [85, 90, "-", 88, 76],
    "Department": ["CSE", "IT", "CSE", "cse", "E&TC"],
    "Admission_Date": ["01-06-2023", "03-06-2023", "05-06-2023", "07-06-2023", "10/06/2023"]
}
df = pd.DataFrame(data)
df
```

```python
##saving original dataset as csv file
df.to_csv("student_data.csv", index=False)

print("\nCleaned CSV file saved as student_data.csv")
```

```python
# Step 2--->Replacing '-' with NaN to markk missing values
df.replace("-", np.nan, inplace=True)

##Converting columns to numeric where needed
df["Age"] = pd.to_numeric(df["Age"], errors='coerce')
df["Marks"] = pd.to_numeric(df["Marks"], errors='coerce')
df

##when converting data like strings-->datatime, if some values are invalid or cannot be converted then instead of showing an error, it will replace those invalid values with NaT(Not a Time)
```

```python
# Step 3---> Check missing values
print("\nMissing Values: ")
print()
print(df.isnull().sum())   ##identifies all null/NaN values and sums it
```

```python
# Step 4--> Fill missing Age and Marks with mean
df["Age"].fillna(df["Age"].mean(), inplace=True)
df
#Fill missing Marks with Median
df["Marks"].fillna(df["Marks"].median(), inplace=True)
df
```

```python
# Step 5--->Fixing inconsistent text formatting
df["Department"]= df["Department"].str.upper()
df
```

```python
# Step 6---> To convert admission date to standard datetime format
df["Admission_Date"] = pd.to_datetime(df["Admission_Date"], errors='coerce')
df
```

```python
# Step 7--->Cleaned Data
print("Cleaned Data: ")
print()
df
```

```python
# Step 8---> save cleaned dataset
df.to_csv("Cleaned_student_data.csv", index=False)

print("\nCleaned CSV file saved as cleaned_student_data.csv")
```
